# Questão 1: 
Um investimento inicial de $1000$ Reais rende juros compostos a uma taxa de 8% ao ano. Faça um gráfico que mostre como o valor do investimento varia com o tempo e determine qual será o valor do investimento após 10 anos. Utilize a equação $V = P * (1 + r)^t$, onde $P$ é o valor principal do investimento, $r$ é a taxa de juros anual (em decimal) e $t$ é o tempo decorrido em anos.


```python
import matplotlib.pyplot as plt
import numpy as np

#codigo para printar os valores do investimento ao longo dos anos

def investimento(p):
  print("investimento com 0 juros de rendimento:1000")
  print("Investimento com 1 juro de rendimento:1080")
  t=2
  while t <= 10:
    v = p*(1 + 0.08)**t
    print(f"Investimento com {t} juros de rendimento:",v)
    t += 1


investimento(1000)

#codigo para plotar o grafico

def calcular_investimento(p, r, t):
    return 1000 * (1 + 0.08) ** t

anos = range(11)

valores_investimento = [calcular_investimento(1000, 0.08, t) for t in anos]

plt.plot(anos, valores_investimento)
plt.xlabel('Anos')
plt.ylabel('Valor do Investimento')
plt.title('Investimento com Juros de Rendimento')
plt.grid(True)
plt.show()


```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x)= e^x - 4x^2$

2. $f(x)= \log_3(x^2 + 1) - \log_3(x) -12$

```python
import numpy as np
import matplotlib.pyplot as plt

#Questão 1

def f(x):
    return np.exp(x) - 4 * x ** 2

x = np.linspace(-5, 5, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = e^x - 4x^2$')

if f(0) < f(1):
    plt.text(-4, 20, 'Função crescente')
else:
    plt.text(-4, 20, 'Função decrescente')

plt.grid(True)
plt.show()


def f(x):
    return np.log10(x**2 + 1)/np.log10(3) - np.log10(x)/np.log10(3) - 1/2

x = np.linspace(0.1, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função $f(x)=\\frac{\\log_{3}(x^2+1)}{\\log_{3}(x)}-\\frac{1}{2}$')

if f(0.1) < f(10):
    plt.text(5, 0, 'Função crescente')
else:
    plt.text(5, 0, 'Função decrescente')
plt.grid(True)
plt.show()
```
